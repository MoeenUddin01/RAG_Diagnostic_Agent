# PlantVillage Class Counts

This notebook counts the number of image files in each class folder under:

- `dataset/raw/PlantVillage`

It produces a per-class table and a total image count.


```python
from __future__ import annotations

from pathlib import Path

import pandas as pd
```

```python
BASE_DIR = Path.cwd().resolve().parent
DATASET_DIR = BASE_DIR / 'dataset' / 'raw' / 'PlantVillage'

assert DATASET_DIR.exists(), DATASET_DIR
DATASET_DIR
```

```python
def count_files_in_class(class_dir: Path) -> int:
    """Count regular files inside a single class directory."""
    return len([path for path in class_dir.iterdir() if path.is_file()])


class_counts = {
    class_dir.name: count_files_in_class(class_dir)
    for class_dir in sorted(DATASET_DIR.iterdir())
    if class_dir.is_dir()
}

counts_df = (
    pd.Series(class_counts, name='image_count')
    .sort_index()
    .rename_axis('class_name')
    .reset_index()
)
counts_df
```

```python
print(f'Total classes: {len(class_counts)}')
print(f'Total images: {counts_df['image_count'].sum():,}')
```

```python
counts_df.sort_values('image_count', ascending=False).reset_index(drop=True)
```
